```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Assignment-Ratings.csv", index_col='User',nrows=20, usecols=range(21))
df = df.fillna(0)
print df.head()
```

```python
L2=np.sqrt(np.square(df[0:20]).sum(axis=0)).to_frame()
print L2
```

```python
norm_df = df[df>0].subtract(df[df>0].mean(axis=1), axis=0).fillna(0)
print norm_df.head()
```

```python
mf = df[0:20].T.dot(df[0:20])

divisor = L2.dot(L2.T)

rating = mf.divide(divisor, axis=0)
print rating.iloc[:20, 0].sort_values(ascending=False)
```

```python
user_rating = df.loc[[5277]]
#df.loc[:, (df != 0).any(axis=0)]
user_sum = rating.loc[ :, (user_rating > 0).any(axis=0)].sum(axis=1)
#print rating[user_rating>0].sum()
(user_rating.dot(rating)/user_sum).T
```

```python
norm_L2=np.sqrt(np.square(norm_df).sum(axis=0)).to_frame()
print norm_L2
```

```python
norm_dot = norm_df.T.dot(norm_df)

norm_div = norm_L2.dot(norm_L2.T)

norm_rating = norm_dot.divide(norm_div, axis=0)
print norm_rating.iloc[:20, 0].sort_values(ascending=False)
```

```python
norm_rating[norm_rating<0]=0
user_norm_sum = norm_rating.loc[ :, (user_rating > 0).any(axis=0)].sum(axis=1)
#print rating[user_rating>0].sum()
(user_rating.dot(norm_rating)/user_norm_sum).T
```
